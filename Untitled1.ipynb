```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("https://github.com/abinesh4106/Sales-analysis/raw/refs/heads/main/sales_data_sample.csv", encoding="latin1")
df.head()
```

```python
column_mapping = {
    "ORDERNUMBER": "Order_ID",
    "QUANTITYORDERED": "Quantity_Ordered",
    "PRICEEACH": "Price_Per_Unit",
    "ORDERLINENUMBER": "Order_Line_Number",
    "SALES": "Sales",
    "ORDERDATE": "Order_Date",
    "STATUS": "Order_Status",
    "QTR_ID": "Quarter",
    "MONTH_ID": "Month",
    "YEAR_ID": "Year",
    "PRODUCTLINE": "Product_Line",
    "MSRP": "MSRP",
    "PRODUCTCODE": "Product_Code",
    "CUSTOMERNAME": "Customer_Name",
    "PHONE": "Phone",
    "ADDRESSLINE1": "Address_Line_1",
    "ADDRESSLINE2": "Address_Line_2",
    "CITY": "City",
    "STATE": "State",
    "POSTALCODE": "Postal_Code",
    "COUNTRY": "Country",
    "TERRITORY": "Territory",
    "CONTACTLASTNAME": "Contact_Last_Name",
    "CONTACTFIRSTNAME": "Contact_First_Name",
    "DEALSIZE": "Deal_Size"
}

df.rename(columns=column_mapping, inplace=True)
df.columns
```

```python
df["Order_Date"] = pd.to_datetime(df["Order_Date"])

# Feature Engineering
df["Revenue"] = df["Sales"]

df["Day_Name"] = df["Order_Date"].dt.day_name()

df["Month_Name"] = df["Order_Date"].dt.month_name()
df['Revenue']
```

```python
print("\nDataset Shape")
print(df.shape)
```

```python
print("\nMissing Values")
print(df.isnull().sum())
```

```python
print("\nSummary Statistics")
print(df.describe())
```

```python
# NumPy Analysis
sales_array = np.array(df["Revenue"])
sales_array
```

```python
print("\nTotal Revenue:", np.sum(sales_array))
print("Average Revenue:", np.mean(sales_array))
print("Highest Sale:", np.max(sales_array))
print("Lowest Sale:", np.min(sales_array))
print("Standard Deviation:", np.std(sales_array))
```

```python
# Product Line Analysis
product_sales = (
    df.groupby("Product_Line")["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

# Country Analysis
country_sales = (
    df.groupby("Country")["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

# Deal Size Analysis
deal_sales = (
    df.groupby("Deal_Size")["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

# Monthly Sales
monthly_sales = (
    df.groupby("Month_Name")["Revenue"]
    .sum()
)

# Top Customers
top_customers = (
    df.groupby("Customer_Name")["Revenue"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
```

```python
print("\nTop Product Lines")
print(product_sales)
```

```python

print("\nTop Countries")
print(country_sales.head(10))
```

```python

print("\nTop Customers")
print(top_customers)
```

```python
# Product Line Revenue
plt.figure(figsize=(12,6))
sns.barplot(
    x=product_sales.index,
    y=product_sales.values
)
plt.xticks(rotation=45)
plt.title("Revenue by Product Line")
plt.tight_layout()
plt.show()
```

```python
# Country Revenue
plt.figure(figsize=(12,6))
sns.barplot(
    x=country_sales.head(10).index,
    y=country_sales.head(10).values
)
plt.xticks(rotation=45)
plt.title("Top 10 Countries by Revenue")
plt.tight_layout()
plt.show()

# Deal Size Distribution
plt.figure(figsize=(8,5))
sns.countplot(
    x="Deal_Size",
    data=df
)
plt.title("Deal Size Distribution")
plt.show()
```

```python
# Monthly Revenue Trend
monthly_sales = monthly_sales.reindex([
    "January","February","March","April",
    "May","June","July","August",
    "September","October","November","December"
]).dropna()

plt.figure(figsize=(10,5))
sns.lineplot(
    x=monthly_sales.index,
    y=monthly_sales.values,
    marker="o"
)
plt.xticks(rotation=45)
plt.title("Monthly Revenue Trend")
plt.tight_layout()
plt.show()
```

```python
# Revenue Distribution
plt.figure(figsize=(8,5))
sns.histplot(
    df["Revenue"],
    bins=30,
    kde=True
)
plt.title("Revenue Distribution")
plt.show()
```

```python
# Correlation Heatmap
numeric_df = df.select_dtypes(include=np.number)

plt.figure(figsize=(10,6))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap="coolwarm"
)
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Top Customers
plt.figure(figsize=(12,6))
sns.barplot(
    x=top_customers.values,
    y=top_customers.index
)
plt.title("Top 10 Customers")
plt.tight_layout()
plt.show()
```

```python
print("\nBusiness Insights")
print(f"Best Product Line : {product_sales.idxmax()}")
print(f"Best Country      : {country_sales.idxmax()}")
print(f"Largest Deal Size : {deal_sales.idxmax()}")
print(f"Top Customer      : {top_customers.idxmax()}")
```
